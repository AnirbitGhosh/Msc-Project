```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
sam_1 = np.array([[0.708, 1.587, 2.417, 2.246], [0.464, 0.781, 1.343, 1.855]])
sam_2 = np.array([[0.269, 0.684, 1.758, 2.905], [0.854, 1.953, 4.492, 7.104]])
sam_3 = np.array([[0.928, 2.026, 4.37, 6.763], [0.781, 1.562, 2.075, 2.979]])
sam_4 = np.array([[0.708, 1.392, 3.076, 4.224], [0.342, 1.099, 1.807, 3.027]])
sam_5 = np.array([[0.61, 1.953, 2.588, 3.174], [0.293, 0.659, 1.245, 2.734]])
sam_6 = np.array([[0.659, 1.685, 3.638, 5.396], [0.269, 1.685, 2.979, 4.028]])
sam_7 = np.array([[0.757, 1.538, 2.271, 2.783], [0.562, 0.879, 1.099, 1.831]])
sam_8 = np.array([[0.708, 0.806, 0.854, 1.147], [0.342, 1.27, 2.417, 3.467]])
sam_9 = np.array([[0.586, 1.367, 2.49, 5.005], [0.317, 0.366, 1.587, 2.319]])
sam_10 = np.array([[0.977, 1.807, 2.441, 2.783], [0.439, 1.147, 2.148, 2.539]])


# Stacking the arrays to calculate the averages across the samples
all_samples = np.array([sam_1, sam_2, sam_3, sam_4, sam_5, sam_6, sam_7, sam_8, sam_9, sam_10])

# Calculating the average of the first array across all samples
average_overestimate = np.mean(all_samples[:, 0, :], axis=0)

# Calculating the average of the second array across all samples
average_underestimate = np.mean(all_samples[:, 1, :], axis=0)

# Printing the results
print("Average of the first array across all samples:")
print(average_overestimate)

print("\nAverage of the second array across all samples:")
print(average_underestimate)
```

```python
def plot_bar_chart(over, under):
    x_values = [30, 60, 90, 120]
    width = 10 # Width of the bars

    # Creating the plot
    fig, ax = plt.subplots(figsize=(15, 10))

    # Bar positions for array_1 and array_2
    bar1 = ax.bar(np.array(x_values) - width/2, over, width, label='Avg overestimate', color='red')
    bar2 = ax.bar(np.array(x_values) + width/2, under, width, label='Avg underestimate', color='blue')

    # Labeling and formatting
    ax.set_xlabel('Time after ignition (hours)', fontsize=16)
    ax.set_ylabel('Mismatch (% of total area)', fontsize=16)
    ax.set_xticks(x_values)
    ax.tick_params(axis='both', labelsize=14)  # Increase font size of tick labels
    ax.legend(fontsize=14)  # Increase legend font size

    # Show the plot
    plt.show()
    
plot_bar_chart(average_overestimate, average_underestimate)
```
